### Vamos a ver el sesgo y la kurtosis

%f es lo que se llama "String formating operator"  . Podemos encontrar más información por ejempl aquí https://www.learnpython.org/en/String_Formatting

```python
print("Sesgo %f" % Amesdataset["SalePrice"].skew())
```

```python
print("Curtosis %f" % Amesdataset["SalePrice"].kurt())
```

Lo que haríamos ahora sería ver si existe una relación entre la variable objetivo y aquellas variables numéricas que consieramos que pueden explicarla. Vamos a ver qué tipo de relación tienen

Aunque se puedene hacer con varias librerías, vamos a hacerla con Seaborn que está muy bien y nos ofrece muchas posibilidades de personalización del gráfico.

Vamos a ver un gráfico de regresión. Tanto Seaborn como mapplotlib están muy biene y tienen  muchísimas opciones para la realización de gráficos. 

```python
sns.lmplot(x="GrLivArea", y="SalePrice",data=Amesdataset,hue="HouseStyle")
```

```python


sns.lmplot(x='TotalBsmtSF',y="SalePrice",data=Amesdataset)
```

Vamos a ver ahora las relaciones con las variables categóricas. Para ello utilizamos los BoxPlot

```python
sns.boxplot(x="OverallQual",y="SalePrice",data=Amesdataset)
```

```python
Fig=sns.boxplot(x="YearBuilt",y="SalePrice",data=Amesdataset)
```

Estos dos gráficos nos indican como se comporta una variable numérica (en este caso el precio) con respecto a una variable categórica (por ejemplo el día de la semana, un modelo de coche, etc)
